### Eliminar columnas 

Creo que para este análisis podemos prescindir de la informacion sobre vuelos cancelados o desviados, el tiempo que pasa el avión en pista con los pasajeros dentro (TaxiIn, TaxiOut), la distancia y el destino del vuelo, el tiempo que ha pasado el avion en el aire, el número de vuelo y tambien la columna del año (todas son 2008) y unnamed.

```python
columnas_para_eliminar = ['Unnamed: 0', 'Year', 'Month', 'DayofMonth', 'DayOfWeek', 
                          'FlightNum', 'Cancelled', 'CancellationCode', 'Diverted', 'TaxiIn', 'TaxiOut']
airlines_delay_reducido = df_sin_nan.drop(columnas_para_eliminar, axis=1)
airlines_delay_reducido
```

### Tipo de dato
Para unificar el dataframe y viendo que todos los decimales son .0, defino una función para covertrlo todo a int64.

```python
def float_a_int(valor):
    if isinstance(valor, float):
        return int(valor)
    else:
        return valor

airlines_delay_reducido = airlines_delay_reducido.applymap(float_a_int)
airlines_delay_reducido
```

```python
columnas_horarias = ['DepTime', 'CRSDepTime', 'ArrTime', 'CRSArrTime']

for columna in columnas_horarias:
    airlines_delay_reducido[columna] = pd.to_datetime(airlines_delay_reducido[columna], format='%H%M', errors='coerce')
for columna in columnas_horarias:
    airlines_delay_reducido[columna] = airlines_delay_reducido[columna].dt.strftime('%H:%M')

airlines_delay_reducido
```

## Muestra aleatoria
Usamos una muestra aleatoria de 200.000 registros. Para asegurar que cada vez que se ejecute el código sea igual, utilizo el metodo random_state para fijar la semilla a partir de la cual genera 'muestra_airlines_delay'. Tenemos resultados numericos del tipo int64 y float64. Lo paso todo a float64 por la mayoria de columnas que ya tiene este tipo de dato.    

```python
muestra_airlines_delay = airlines_delay_reducido.sample(n=200000, random_state=42)
muestra_airlines_delay
```

#### Fes una taula de les aerolínies amb més endarreriments acumulats.

Conteo de vuelos por cada aerolinea. A continuación se muestra el número de retrasos de cada una de ellas y el porcentaje.

```python
num_aerolineas = muestra_airlines_delay['UniqueCarrier'].value_counts()
num_retrasos = muestra_airlines_delay[muestra_airlines_delay['Delayed'] > 0].groupby('UniqueCarrier')['Delayed'].count()
resultados_combinados = pd.DataFrame({'VuelosTotales': num_aerolineas, 'Retrasos': num_retrasos})
resultados_combinados['PorcentajeRetrasos'] = (resultados_combinados['Retrasos'] / resultados_combinados['VuelosTotales']) * 100
resultados_combinados = resultados_combinados.sort_values(by='VuelosTotales', ascending=False)

print(resultados_combinados)
```

#### Quins són els vols més llargs? I els més endarrerits? Busca les rutes més llargues i les que acumulen més retards.

```python
rutas = muestra_airlines_delay.sort_values(by='Distance', ascending=False)
rutas_mas_largas = rutas[['Origin', 'Dest', 'Distance']].drop_duplicates(subset='Distance')

print(rutas_mas_largas)
```

La ruta más larga es de Newark (EWR) a Honolulu (HNL).

```python
retrasos_acumulados = muestra_airlines_delay.groupby(['Origin', 'Dest'])['Delayed'].sum().reset_index()
rutas_con_mas_retrasos = retrasos_acumulados.sort_values(by='Delayed', ascending=False)

print(rutas_con_mas_retrasos)
```

La ruta que más retrasos acumula es Los Angeles (LAX) San Francisco (SFO).
