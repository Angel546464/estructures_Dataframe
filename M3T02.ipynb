Selecciono las filas con valores nulos en AirTime (tiempo que el avión pasa en el aire) y ArrDelay (retraso sobre el tiempo previsto de llegada del avión).

```python
filas_nan = airlines_delay[airlines_delay['AirTime'].isnull() & airlines_delay['ArrDelay'].isnull()]
filas_nan
```

```python
cancelados_o_desviados = filas_nan[(filas_nan['Cancelled'] == 1) | (filas_nan['Diverted'] == 1)]
cancelados_o_desviados
```

Coinciden los vuelos cancelados o desviados con los que no tienen datos en AirTime y en ArrDelay. Elimino estas filas porque al ser vuelos cancelados o desviados pueden producir valores atípicos en los cálculos.

```python
airlines_delay_no_nan = airlines_delay.drop(filas_nan.index)
airlines_delay_no_nan
```

Con esta operación se eliminan 8387 entradas, representando el 0.43% del total de la muestra y sabemos que los vuelos que quedan no han sido cancelados o desviados.

```python
print (airlines_delay_no_nan.isnull().sum())
```

Al eliminar cancelados y desviados desaparecen NaN en casi todas las categorías excepto las que clasifican de quién es consecuencia el retraso. 
También hay un valor de 3 NaN en TailNum (identificador único del avión). 
Para eliminar todos los datos faltantes hay que suprimir 680883 entradas lo que representa el 35.16% del total del conjunto. Es una cantidad considerable, pero la falta de información de las causas del retraso podría llevar a un análisis incompleto.

```python
airlines_delay_no_nan = airlines_delay_no_nan.dropna()
airlines_delay_no_nan
```

### ELIMINAR COLUMNAS
Creo que para este análisis podemos prescindir de la informacion sobre vuelos cancelados o desviados, el tiempo que pasa el avión en pista con los pasajeros dentro (TaxiIn, TaxiOut), la distancia y el destino del vuelo, el tiempo que ha pasado el avion en el aire, el número de vuelo y tambien la columna del año (todas son 2008) y unnamed.

```python
columnas_para_eliminar = ['Unnamed: 0', 'Year', 'Month', 'DayofMonth', 'DayOfWeek', 
                          'FlightNum', 'Cancelled', 'CancellationCode', 'Diverted', 'TaxiIn', 'TaxiOut']
airlines_delay_no_nan = airlines_delay_no_nan.drop(columnas_para_eliminar, axis=1)
airlines_delay_no_nan
```

### TIPO DE DATO
Primero compruebo qué tipos de dato contiene el dataset.

```python
tipos_de_datos = airlines_delay_no_nan.dtypes
print(tipos_de_datos)
```

```python
def float_a_int(valor):
    if isinstance(valor, float):
        return int(valor)
    else:
        return valor

airlines_delay_no_nan = airlines_delay_no_nan.applymap(float_a_int)
airlines_delay_no_nan
```

```python
columnas_horarias = ['DepTime', 'CRSDepTime', 'ArrTime', 'CRSArrTime']

for columna in columnas_horarias:
    airlines_delay_no_nan[columna] = pd.to_datetime(airlines_delay_no_nan[columna], format='%H%M', errors='coerce')
for columna in columnas_horarias:
    airlines_delay_no_nan[columna] = airlines_delay_no_nan[columna].dt.strftime('%H:%M')

# Ahora airlines_delay tiene nuevas columnas que contienen solo la hora
airlines_delay_no_nan
```

```python
valores_nulos_por_columnas = airlines_delay_no_nan.isnull().sum()
print (valores_nulos_por_columnas)
```

Tras la conversión a formato datetime se vuelven a generar valores nulos.

```python
airlines_delay_def=airlines_delay_no_nan.dropna()
airlines_delay_def
```

### MUESTRA ALEATORIA

Usamos una muestra aleatoria de 200.000 registros. Para asegurar que cada vez que se ejecute el código sea igual, utilizo el metodo random_state para fijar la semilla a partir de la cual genera 'muestra_airlines_delay'.

### Fes una taula de les aerolínies amb més endarreriments acumulats.
Conteo de vuelos por cada aerolinea. A continuación se muestra el número de retrasos de cada una de ellas y el porcentaje. En la primera tabla se ordenan por número de vuelos de mayor a menor, y en la segunda tabla se muestran ordenadas por el porcentaje de retrasos de menor a mayor.

```python
num_aerolineas = muestra_airlines_delay['UniqueCarrier'].value_counts()
num_retrasos = muestra_airlines_delay[muestra_airlines_delay['Delayed'] > 0].groupby('UniqueCarrier')['Delayed'].count()
retrasos_aerolineas = pd.DataFrame({'VuelosTotales': num_aerolineas, 'Retrasos': num_retrasos})
retrasos_aerolineas['PorcentajeRetrasos'] = (retrasos_aerolineas['Retrasos'] / retrasos_aerolineas['VuelosTotales']) * 100
retrasos_aerolineas = retrasos_aerolineas.sort_values(by='VuelosTotales', ascending=False)

retrasos_aerolineas
```

```python
retrasos_aerolineas = retrasos_aerolineas.sort_values(by='PorcentajeRetrasos', ascending=True)
retrasos_aerolineas
```

```python
retrasos = retrasos_aerolineas['PorcentajeRetrasos'].describe()
retrasos
```

### Quins són els vols més llargs? I els més endarrerits? Busca les rutes més llargues i les que acumulen més retards.

```python
rutas = muestra_airlines_delay.sort_values(by='Distance', ascending=False)
rutas_mas_largas = rutas[['Origin', 'Dest', 'Distance']].drop_duplicates(subset='Distance')

print(rutas_mas_largas)
```

La ruta más larga es de Newark (EWR) a Honolulu (HNL) con 4962 millas.

```python
retrasos_acumulados = muestra_airlines_delay.groupby(['Origin', 'Dest'])['Delayed'].sum().reset_index()
rutas_con_mas_retrasos = retrasos_acumulados.sort_values(by='Delayed', ascending=False)

print(rutas_con_mas_retrasos)
```

Los trayectos que más retrasos acumulan son Los Angeles (LAX) San Francisco (SFO) con 494 retrasos, después viene el trayecto Chicago (ORD) New York (LAG) con 488 vuelos, en tercera posición vendría el mismo trayecto ubicado en primera posición, pero al revés (SFO)-(LAX) y el cuarto es el que cubre el trayecto Atlanta (ATL) New York (LGA) con 431 retrasos acumulados.

Son aeropuertos muy transitados. El de Atlanta (ATL) es el de más tráfico aéreo del mundo, el cuarto en ese ranking es el de Chicago (ORD), y Los Angeles (LAX) se encuentra en la sexta posición.

Con lo que de los cinco aeropuertos implicados en los cuatro trayectos que más retrasos acumulan en nuestra muestra, tres de ellos están entre los 10 más transitados del mundo. Creo que este puede ser uno de los motivos de que tengan más retrasos.
