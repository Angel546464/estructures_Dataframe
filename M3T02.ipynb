## Datos y variables de la muestra
Dataframe con información sobre vuelos en EE. UU. durante 2008 con datos recopilados por la Oficina de Estadísticas de Transporte (BTS) del Departamento de Transporte de EE. UU. (DOT).

El data set contempla hasta 30 variables sobre 1.936.758 millones de entradas de las que nos quedamos con una muestra aleatoria de 200000 para el análisis. Cada entrada o fila representa los datos de un vuelo en concreto incluyendo la información del avión, fechas, horarios esperados y reales, número de vuelo, origen, destino, distancia, tiempos de vuelo, los que se han cancelado o desviado, los que se han retrasado y las causas de esos retrasos.
